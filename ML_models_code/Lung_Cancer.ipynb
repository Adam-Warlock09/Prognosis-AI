```python
import numpy as np
import pandas as pd
print("Setup Complete")
```

```python
cancer_file_path = '/content/Lung_Cancer_Dataset.csv'
cancer_data = pd.read_csv(cancer_file_path)
cancer_data.columns
```

```python
cancer_data.head(10)
```

```python
cancer_data = cancer_data.dropna(axis=0)

cancer_data.LUNG_CANCER = [int(x == "YES") for x in cancer_data.LUNG_CANCER]
cancer_data.GENDER = [int(x == "M") for x in cancer_data.GENDER]
# print(cancer_data.LUNG_CANCER)
# print(cancer_data.GENDER)
```

```python
y = cancer_data.LUNG_CANCER
y.describe()
```

```python
cancer_features =cancer_features = [ 'GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY', 'ALLERGY ', 'ALCOHOL CONSUMING', 'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY', 'CHEST PAIN', 'CHRONIC DISEASE', 'WHEEZING']
cancer_data.AGE.describe()
```

```python
X = cancer_data[cancer_features]
```

```python
X.describe()
```

```python
from sklearn.model_selection import train_test_split

train_x , val_x , train_y , val_y =  train_test_split( X, y, test_size = 0.2, random_state = 0) #destructuring

from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

forest_model = RandomForestRegressor( n_estimators= 350 , random_state = 1 )

forest_model.fit(train_x, train_y)

cancer_preds = forest_model.predict(val_x)
# cancer_preds.head()

print(mean_absolute_error(val_y,cancer_preds))
```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor( n_estimators = 450 , learning_rate = 0.035)
xgb_model.fit(train_x , train_y)

xgb_preds = xgb_model.predict(val_x)

print(mean_absolute_error(val_y, xgb_preds))
```
